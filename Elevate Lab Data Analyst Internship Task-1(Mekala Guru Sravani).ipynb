```python
# STEP 4: Import Pandas and Read the Dataset
import pandas as pd

df = pd.read_csv("netflix_titles.csv.csv")

df.head()
```

```python
# STEP 5: Basic Structure Analysis
df.shape

# Column Names:
print("Column Names: ")
df.columns
print()

# Data Types
print("Data Types: ")
df.info()
```

```python
# STEP 6: Check Missing Values Properly
df.isnull().sum()
```

```python
# STEP 7: Handle Missing Values

# director column (2634 missing)
df['director'] = df['director'].fillna("Unknown")
# Now verify director column again:
df['director'].isnull().sum()



# cast (825 missing) column:
df['cast'] = df['cast'].fillna("Not Available")
#  Now verify cast column again:
df['cast'].isnull().sum()



# country (831 missing) column:
df['country'] = df['country'].fillna("Unknown")
#  Now verify country column again:
df['country'].isnull().sum()

```

```python
# director, cast, and country columns should show zero
df.isnull().sum()
```

```python
# rating (4 missing) column:
# Very small number. We can fill with most frequent rating (mode).
df['rating'].mode()
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

```python
# now rating also shows zero
df.isnull().sum()
```

```python
# duration (3 missing) column:
# Only 3 rows.Best option: Drop those rows.
df = df.dropna(subset=['duration'])

# Check shape again:
df.shape # now it should show only 8804 rows earlier 8807 rows.
```

```python
# date_added (10 missing) column:
# We can drop them since very few.
df = df.dropna(subset=['date_added'])

# Check shape again:
df.shape
```

```python
# Final Verification
df.isnull().sum() # there should not be any null values now.. everyhting should be zero..
```

```python
# STEP 8: Check for Duplicate Records
df.duplicated().sum()
```

```python
# STEP 9: Standardize Text Columns

# Remove Extra Spaces from All Object Columns
# Select all object (string) columns
string_columns = df.select_dtypes(include='object').columns

# Remove leading and trailing spaces
for col in string_columns:
    df[col] = df[col].str.strip()



# Standardize Column Names
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.columns
```

```python
# STEP 10: Convert date_added to Proper Date Format
df['date_added'] = pd.to_datetime(df['date_added'])
df.info()
```

```python
# STEP 11: Create Separate Year & Month
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
df.info()
```

```python
# STEP 12: Clean the duration Column
# Extract number
df['duration_number'] = df['duration'].str.extract(r'(\d+)').astype(int)

# Create structured columns
df['duration_minutes'] = df.apply(
    lambda x: x['duration_number'] if x['type'] == 'Movie' else 0,
    axis=1
)

df['duration_seasons'] = df.apply(
    lambda x: x['duration_number'] if x['type'] == 'TV Show' else 0,
    axis=1
)

# Drop temporary columns
df = df.drop(columns=['duration', 'duration_number'])
```
